### Tensor datatypes - Tensor attribute

Aside from ***shape issues*** (tensor shapes don't match up), two of the other most common issues you'll come across in PyTorch are ***datatype*** and ***device issues.***

1. Tensors not right datatype -> tensor.dtype
2. Tensors not right shape -> tensor.shape
3. Tensors not on the right device -> tensor.device

```python
# Float 32 tensor => default tensor datatype is float32
# float_32_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=torch.float16)
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None, # defaults to None, which is torch.float32 or whatever datatype is passed
                               device=None, # defaults to None, which uses the default tensor type (cpu or cuda(gpu))
                               requires_grad=False) # if True, operations performed on the tensor are recorded
float_32_tensor, float_32_tensor.dtype
```

```python
# float_16_tensor = torch.tensor([3.0, 6.0, 9.0], dtype=torch.float16)
float_16_tensor = float_32_tensor.type(torch.float16)
float_16_tensor
```

```python
result = float_16_tensor * float_32_tensor
result, result.dtype
```

### Reshaping, stacking, sqeezing and unsquezzing tensors

* Reshaping -> reshapes an input tensor to a define shape
* View -> Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking -> combine multiple tensors on top of each other (vstack) or side by side (hstack)
*Squeeze - remove all '1' dimentions from a tensor
* Unsqueeze - add a '1' dimention to a target tensor
* Permute - Return a view of the input with permuted (swapped) in a certain way

```python
# Let's create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimentions
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

Changing z -> changing x (because a view of a tensor shares the same memory as the original input)

```python
z[:, 0] = 1
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
x_reshaped, x_reshaped.shape
```

```python
# Squeeze -> remove all single dimention
x_squeeze = torch.squeeze(x_reshaped)
x_squeeze, x_squeeze.shape, x_reshaped.squeeze().shape
```

```python
# torch.unsqueeze() -> add a single dimention to a target tensor at a specific dimention
print(f"Previous target: {x_squeeze}")
print(f"Previous shape: {x_squeeze.shape}")

# Add an extra dimention with unsqueeze
x_unsqueezed = x_squeeze.unsqueeze(dim=0)
print('Unsqueeze: ')
x_unsqueezed, x_unsqueezed.shape
```

```python
# torch.permute - rearranges the dimentions of a target tensor in a specify order
x_original = torch.rand(size=(224, 224, 3))  # [height, width, color_channel]

# Permute the original tensor to rearrange the axis ot dims order
# 2 =>color_channel, 0 => width, 1 => height
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # [color_channels, height, width]
```

One of the most common errors in deep learning: shape

```python
x_original[0, 0, 0] = 123456
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## Reproducbility (trying to take random out of random)

In short how a neural network learns:

`start with random number -> tensor operations -> update random numbers to try and make them better representations of the date -> agai -> again -> again...`

to reduce the randoness in neural network and Pytorch comes the concept of a **random seed**

Essentially what the random seed does is "flavour" the randoness

```python
import torch

# Create two random tensor
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
#  Let's make some random but reproducible tensors
import torch

#  set the random seed
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```

## 3. Putting tensors (and models) on the GPU

The reason we want our tensors/models on the GPU is because using GPU results is faster for computations

```python
# create a tensor (default on the CPU)
tensor = torch.tensor([1, 2, 3])
print(tensor, tensor.device)
```

```python
# move tensor if available
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
# result :tensor([1, 2, 3], device='cuda:0') => 0 is an index for cuda device that we use
```

**Numpy only work with CPU** => we need to move tensors back to CPU

### 4. Moving tensors back to the CPU


```python
# if tensors is on GPU, can't transform it to numpy
# tensor_on_gpu.numpy() -> this will cause an error as the device is wrong!!!
# Error: can't convert cuda:0 device type tensor to numpy. Use Tensor.cpu() to copy the tensor to host memory first.
```

```python
# to fix the GPU tensor with umPy issue, we can first set it to the CPU
tensor_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_on_cpu
```
